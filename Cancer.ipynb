```python
import seaborn as sns
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['id'],axis=1)
```

```python
df=df.drop(['Unnamed: 32'],axis=1)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
df.head()
```

```python
#as some values have very large values like area_mean and some have
#very small values such as smoothness_mean 
#so we need ot normalize the data using StandardScaler
#: Scaling ensures that each feature contributes equally to the model
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
  x = df.drop(['diagnosis'], axis=1)
  y = df['diagnosis']
```

```python
scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Train the model
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
model = LogisticRegression()
```

```python
 model.fit(x_train, y_train)
```

```python
# test model
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_pred = model.predict(x_test)
```

```python
print('Accuracy of our model: ', accuracy_score(y_test, y_pred))
```

```python
print("Classification report: \n", classification_report(y_test, y_pred))
```
